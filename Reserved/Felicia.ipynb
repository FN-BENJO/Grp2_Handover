```python
#data cleaning for combine_file

#create a copy of dataset for data cleaning
cleaned_combine_file = combine_file.copy()

#rename variables with more representative names
cleaned_combine_file.rename(columns= {'Volume_1' : 'Vol_1st_med', 'Volume_2': 'Vol_2nd_med', 'Volume_3': 'Vol_3rd_med'} , inplace= 'True')

#merge variables together
# Disposible? and Disposible variables
cleaned_combine_file['Disposable?'] = np.where(cleaned_combine_file['Disposable'].replace('',np.nan).notna(),cleaned_combine_file['Disposable'],cleaned_combine_file['Disposable?'])
cleaned_combine_file['Disposable?'] = np.where(cleaned_combine_file['Diposable?'].replace('',np.nan).notna(),cleaned_combine_file['Diposable?'],cleaned_combine_file['Disposable?'])

#convert variables into binary variables
cleaned_combine_file['SpecificConsultant?'] = np.where(cleaned_combine_file['Consultant'].isnull(), 'No', 'Yes')
cleaned_combine_file['Disposable?'] = np.where(cleaned_combine_file['Disposable?'].isnull(), 'No', 'Yes')
cleaned_combine_file['Urgency?'] = np.where(cleaned_combine_file['Urgency'].isnull(), 'No', 'Yes')
cleaned_combine_file['Brand'] = np.where(cleaned_combine_file['Brand '].replace('',np.nan).notna(),cleaned_combine_file['Brand '],cleaned_combine_file['Brand'])

```

```python
#ApplicationMethod variable - merge multiple original variables together
merge_list = ['Delivery', 'Delivery Medium', 'Route', 'ROUTE', 'USAGE TYPE']

cleaned_combine_file['ApplicationMethod'] = cleaned_combine_file.loc[:, 'Delivery']
cleaned_combine_file.head()

for col in merge_list:
    selected_col = cleaned_combine_file[col]
    cleaned_combine_file['ApplicationMethod'] = np.where(selected_col.replace('',np.nan).notna(),selected_col, cleaned_combine_file['ApplicationMethod'])

print(cleaned_combine_file['ApplicationMethod'].notna().sum())

merge_list1 = ['Body Part', 'Location']

cleaned_combine_file['OperationArea'] = cleaned_combine_file.loc[:, 'For Body Part']
cleaned_combine_file.head()

for col in merge_list1:
    selected_col = cleaned_combine_file[col]
    cleaned_combine_file['OperationArea'] = np.where(selected_col.replace('',np.nan).notna(),selected_col, cleaned_combine_file['OperationArea'])

print(cleaned_combine_file['OperationArea'].notna().sum())
```

```python
get_excel_file = 'cancer_list.xlsx'
cancer_list = pd.read_excel(get_excel_file)
cancer_list.head()
```

```python
sub_cancer_list = cancer_list.copy()
cols = ['Cancer type', 'Active ingredient']
sub_cancer_list = sub_cancer_list[cols]
sub_cancer_list.head()
```

```python
sub_cancer_list = cancer_list.copy()
cols = ['Cancer type', 'Active ingredient']
sub_cancer_list = sub_cancer_list[cols]

sub_cancer_list.dropna(subset=['Cancer type'], inplace=True)
sub_cancer_list.head()
```

```python
cancer_word_list = []

for cancer_type, ingredient in sub_cancer_list.itertuples(index=False):
    if cancer_type not in cancer_word_list:
        cancer_word_list.append(cancer_type)
    elif ingredient not in cancer_word_list:
        cancer_word_list.append(ingredient)

cancer_word_list
```

```python
cleaned_cancer_List = [x for x in cancer_word_list if str(x) != 'nan']
cleaned_cancer_List
```

```python
cancer_keywords = []

for word in cleaned_cancer_List:
    if word.lower() != 'others':
        get_word = word.split()
        for i in get_word:
            if i not in cancer_keywords:
                cancer_keywords.append(i)

cancer_keywords
```
